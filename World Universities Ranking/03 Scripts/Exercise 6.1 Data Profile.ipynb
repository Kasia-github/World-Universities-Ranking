## 04. Data cleaning

```python
# Check missing values
missing_values = df_CWUR.isnull().sum()
```

```python
print("Missing values per column:")
print(missing_values)
```

```python
df_CWUR['broad_impact'].value_counts(dropna = False)
```

```python
df_CWUR.isnull().sum().sum()
```

```python
# Handling missing values
```

```python
# Calculate the mean value of the 'broad_impact' column
mean_value = df_CWUR['broad_impact'].mean()
```

```python
# Replacing missing values within 'broad_impact' column with the mean value
df_CWUR['broad_impact'].fillna(mean_value, inplace=True)
```

```python
# Check for missing values
df_CWUR.isnull().sum().sum()
```

```python
# Check mixed-type columns
for col in df_CWUR.columns.tolist():
    if len(df_CWUR[col].apply(type).unique())>1:
        print(col+' have mix datatype')
    else:
        print(col+' don\'t have mix datatype')
```

```python
# Check duplicates
df_dups = df_CWUR[df_CWUR.duplicated()]
```
